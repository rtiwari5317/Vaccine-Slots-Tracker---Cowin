```python
import requests
import time
import json
from pygame import mixer
from datetime import datetime,timedelta
from io import StringIO
import sys


age = 30
district_ids = [265,294,670,664]   #Bangalore
#district_ids = [670]    #Lucknow

num_days = 5
print_flag = 'Y'

print("Starting search for Covid vaccine slots!")

actual = datetime.today()
list_format = [actual + timedelta(days=i) for i in range(num_days)]
actual_dates = [i.strftime("%d-%m-%Y") for i in list_format]

while True:
    counter = 0   

     
    for district in district_ids:
        for given_date in actual_dates:
                URL = "https://cdn-api.co-vin.in/api/v2/appointment/sessions/public/findByDistrict?district_id={}&date={}".format(district, given_date)
                header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.76 Safari/537.36'} 

                result = requests.get(URL, headers=header)

                if result.ok:
                    response_json = result.json()
                    if response_json["sessions"]:
                        if(print_flag.lower() =='y'):
                            for session in response_json["sessions"]:
                                if (session["district_name"] == 'Lucknow' and session["min_age_limit"] < age and session["available_capacity_dose1"] > 0 ):
                                        result = StringIO()
                                        sys.stdout = result
                                        print('State: ' + session["state_name"])
                                        print('District: ' + session["district_name"])
                                        print('Pincode: ' + str(session["pincode"]))
                                        print('Age Allowed: '+ str(session['min_age_limit']))
                                        print("Available on: {}".format(given_date))
                                        print("Center Name: ", session["name"])
                                        print("Center Block Name: ", session["block_name"])
                                        print("Price Type: ", session["fee_type"])
                                        print("Availablity : ", session["available_capacity"])
                                        print("Availablity Dose-1 : ", session["available_capacity_dose1"])
                                        print("Availablity Dose-2 : ", session["available_capacity_dose2"])
                                        

                                        if(session["vaccine"] != ''):
                                            print("Vaccine type: ", session["vaccine"])
                                        print("\n")
                                        counter = counter + 1
#                                         mixer.init()
#                                         mixer.music.load('C:\\Users\\U6059833\\Downloads\\sound\\sound_dingdong.wav')
#                                         mixer.music.play()
                                        
                                        #message_string = 'State: ' + session["state_name"] + ' District: ' + session["district_name"] + ' Pincode: ' + str(session["pincode"]) + ' Age Allowed: '+ str(session['min_age_limit']) + " Available on: {}".format(given_date) + " Center Name: ", session["name"] + " Center Block Name: ", session["block_name"] + " Price Type: ", session["fee_type"] + " Availablity : ", str(session["available_capacity"]) + " Availablity Dose-1 : ", str(session["available_capacity_dose1"]) + " Availablity Dose-2 : ", str(session["available_capacity_dose2"]) + "  Vaccine type: ", session["vaccine"]
                                        #txt = str(message_string)
                                        
                                        result_string = result.getvalue()
                                        print(result_string)                                       
                                        if len(result_string) > 0:
                                            base_url = 'https://api.telegram.org/botYOURBOTAPI/sendMessage?chat_id=YOURCHANNELINFO&text={0}'.format(result_string)
                                            print("Response:",requests.get(base_url))
                                            print("Message Sent for 18+!")
                                
                                elif (session["district_name"] == 'Kanpur Nagar' and session["min_age_limit"] < age and session["vaccine"] == 'COVAXIN' and session["available_capacity_dose1"] > 0):
                                        result = StringIO()
                                        sys.stdout = result
                                        print('State: ' + session["state_name"])
                                        print('District: ' + session["district_name"])
                                        print('Pincode: ' + str(session["pincode"]))
                                        print('Age Allowed: '+ str(session['min_age_limit']))
                                        print("Available on: {}".format(given_date))
                                        print("Center Name: ", session["name"])
                                        print("Center Block Name: ", session["block_name"])
                                        print("Price Type: ", session["fee_type"])
                                        print("Availablity : ", session["available_capacity"])
                                        print("Availablity Dose-1 : ", session["available_capacity_dose1"])
                                        print("Availablity Dose-2 : ", session["available_capacity_dose2"])
                                        

                                        if(session["vaccine"] != ''):
                                            print("Vaccine type: ", session["vaccine"])
                                        print("\n")
                                        counter = counter + 1
                                        mixer.init()
                                        #mixer.music.load('C:\\Users\\U6059833\\Downloads\\sound\\sound_dingdong.wav')
                                        #mixer.music.play()
                                        
                                        #message_string = 'State: ' + session["state_name"] + ' District: ' + session["district_name"] + ' Pincode: ' + str(session["pincode"]) + ' Age Allowed: '+ str(session['min_age_limit']) + " Available on: {}".format(given_date) + " Center Name: ", session["name"] + " Center Block Name: ", session["block_name"] + " Price Type: ", session["fee_type"] + " Availablity : ", str(session["available_capacity"]) + " Availablity Dose-1 : ", str(session["available_capacity_dose1"]) + " Availablity Dose-2 : ", str(session["available_capacity_dose2"]) + "  Vaccine type: ", session["vaccine"]
                                        #txt = str(message_string)
                                        
                                        result_string = result.getvalue()
                                        print(result_string)                                       
                                        if len(result_string) > 0:
                                            base_url = 'https://api.telegram.org/botYOURBOTAPI/sendMessage?chat_id=YOURCHANNELINFO&text={0}'.format(result_string)
                                            print("Response:",requests.get(base_url))
                                            print("Message Sent for 18+!")
                                
                                
                                elif (session["min_age_limit"] < age and session["vaccine"] == 'COVAXIN'  and  session["available_capacity_dose1"] > 0 ):
                                        result = StringIO()
                                        sys.stdout = result
                                        print('State: ' + session["state_name"])
                                        print('District: ' + session["district_name"])
                                        print('Pincode: ' + str(session["pincode"]))
                                        print('Age Allowed: '+ str(session['min_age_limit']))
                                        print("Available on: {}".format(given_date))
                                        print("Center Name: ", session["name"])
                                        print("Center Block Name: ", session["block_name"])
                                        print("Price Type: ", session["fee_type"])
                                        print("Availablity : ", session["available_capacity"])
                                        print("Availablity Dose-1 : ", session["available_capacity_dose1"])
                                        print("Availablity Dose-2 : ", session["available_capacity_dose2"])
                                        

                                        if(session["vaccine"] != ''):
                                            print("Vaccine type: ", session["vaccine"])
                                        print("\n")
                                        counter = counter + 1
                                        mixer.init()
                                        #mixer.music.load('C:\\Users\\U6059833\\Downloads\\sound\\sound_dingdong.wav')
                                        #mixer.music.play()
                                        
                                        #message_string = 'State: ' + session["state_name"] + ' District: ' + session["district_name"] + ' Pincode: ' + str(session["pincode"]) + ' Age Allowed: '+ str(session['min_age_limit']) + " Available on: {}".format(given_date) + " Center Name: ", session["name"] + " Center Block Name: ", session["block_name"] + " Price Type: ", session["fee_type"] + " Availablity : ", str(session["available_capacity"]) + " Availablity Dose-1 : ", str(session["available_capacity_dose1"]) + " Availablity Dose-2 : ", str(session["available_capacity_dose2"]) + "  Vaccine type: ", session["vaccine"]
                                        #txt = str(message_string)
                                        
                                        result_string = result.getvalue()
                                        print(result_string)                                       
                                        if len(result_string) > 0:
                                            base_url = 'https://api.telegram.org/botYOURBOTAPI/sendMessage?chat_id=YOURCHANNELINFO&text={0}'.format(result_string)
                                            print("Response:",requests.get(base_url))
                                            print("Message Sent for 18+!")
                else:
                    print("No Response!")
                
    if counter:
        print("No Vaccination slot available!")
    else:
        print("Search Completed! But Not Available")

    dt = datetime.now() + timedelta(seconds = 2)

    while datetime.now() < dt:
        time.sleep(1)
```
